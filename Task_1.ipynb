```python
# 1. Import Libraries & Load Dataset
import pandas as pd

df = pd.read_csv(r"C:\Users\parip\Downloads\archive (3)\netflix_titles.csv")
df.head()
```

```python
# 2. Basic Data Inspection
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
# 3. Handling Missing Values
# Check missing values
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
```

```python
df['cast'] = df['cast'].fillna('Not Available')
```

```python
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
```

```python
df = df.dropna(subset=['date_added'])
```

```python
# Fill or drop based on column importance
# Fill categorical missing values
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)

# Drop rows if critical data missing
df.dropna(subset=['date_added'], inplace=True)
```

```python
# 4. Removing Duplicates
df.drop_duplicates(inplace=True)
```

```python
# 5. Standardizing Text Data
df['country'] = df['country'].str.strip().str.title()
df['type'] = df['type'].str.lower()
```

```python
#  6. Fixing Date Format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
# 7. Renaming Columns (Clean Headers)
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# 8. Fixing Data Types
df['release_year'] = df['release_year'].astype(int)
```

```python
# 9. Outlier Check
df['release_year'].describe()
```

```python
# 10. Final Clean Dataset
df.head()
df.info()
```

```python
df.to_csv("cleaned_netflix_dataset.csv", index=False)
```
